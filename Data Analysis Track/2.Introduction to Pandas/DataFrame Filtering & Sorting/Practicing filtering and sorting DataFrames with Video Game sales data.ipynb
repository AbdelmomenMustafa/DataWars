### Activities

##### 1. Filter Games by Genre

```python
# try your code here
sports_df = df.loc[df["Genre"]=='Sports']
sports_df.head()
```

##### 2. Filter Games by Sales in North America

```python
# Try your code here
na_sales_df = df.loc[df['NA_Sales']>1]
na_sales_df.head()
```

##### 3. Filter Games by Platform and Year

```python
# Try your code here
ps2_2005_df = df.query("Platform == 'PS2' and Year == 2005")
ps2_2005_df.head()
```

##### 4. Filter Games by Global Sales Range

```python
# Try your code here
global_sales_df = df.loc[df['Global_Sales'].between(20.0,30.0)]
global_sales_df.head()
```

##### 5. Filter Games by Publisher and Genre

```python
# Try your code here
nintendo_action_df = df.query("Publisher == 'Nintendo' and Genre == 'Action'")
nintendo_action_df.head()
```

##### 6. Filter Games by Multiple Genres

```python
# Try your code here
action_sports_df = df.loc[(df['Genre']=='Action') | (df['Genre']=='Sports')]
action_sports_df.head()
```

##### 7. Select Games by Multiple Criteria

```python
df.loc[df['Genre'].isin(['Sports','Action'])]
```

```python
# Try your code here

ps2_action_sports_df = df.loc[((df['Genre'].isin(['Sports','Action'])) & 
                              (df['Platform']=='PS2'))&
                               (df['Global_Sales'].between(5.0,30.0))]
ps2_action_sports_df.head()
```

##### 8. Filter Games by Sales

```python
# Try your code here

filtered_query_df = df.loc[df['Global_Sales'].between(5.0,30.0)]
filtered_query_df.head()
```

##### 9. Filter Games by Platform and Critic Score

```python
# Try your code here

query_ps2_df = df.loc[(df['Global_Sales']>5.0) & (df['Platform']=='PS2')]
query_ps2_df.head()
```

##### 10. Filtering Rows and Columns

##### 16. Find all the games released between 2005 and 2010 (inclusive) with multiple condition

```python
# Try your code here
games_2005_2010 = df.query("(Year.between(2005,2010)) and (Genre == 'Action') and (Global_Sales > 5)")
games_2005_2010.head()
```

##### 17. Find the top 10 games with multiple filters

```python
# Try your code here
shooter_2010=df.loc[(df['Genre']=='Shooter') & (df['Year']>2010),['Name', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
top_10_shooter_2010 = shooter_2010.sort_values(by='NA_Sales',ascending=False).head(10)
top_10_shooter_2010
```
